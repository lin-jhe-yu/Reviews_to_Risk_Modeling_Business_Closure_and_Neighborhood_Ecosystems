```python
from pyspark.sql.functions import when, col
combined_df = combined_df.withColumn(
    "irs_estimated_population",
    col("irs_estimated_population").cast("double")
)

# Add permanent closed flag
combined_df = combined_df.withColumn(
    "permanent_closed",
    when(col("state").like("%Permanently closed%"), 1).otherwise(0)
)
```

## 2. EDA

```python
from pyspark.sql.functions import when, col

combined_df_numeric = combined_df.withColumn(
    "price_numeric",
    when(col("price") == "$", 1)
    .when(col("price") == "$$", 2)
    .when(col("price") == "$$$", 3)
    .when(col("price") == "$$$$", 4)
    .when(col("price") == "₩", 1)
    .when(col("price") == "₩₩", 2)
    .when(col("price") == "₩₩₩", 3)
    .when(col("price") == "₩₩₩₩", 4)
    .otherwise(0)
)
```

```python
numeric_cols = [
    "avg_rating",
    "num_of_reviews",
    "latitude",
    "longitude",
    "price_numeric",
    "irs_estimated_population",
    "permanent_closed"
]

from pyspark.ml.feature import VectorAssembler

assembler = VectorAssembler(inputCols=numeric_cols, outputCol="features")
numeric_df = assembler.transform(combined_df_numeric).select("features")
```

```python
from pyspark.ml.stat import Correlation

# Pearson correlation
corr_matrix = Correlation.corr(numeric_df, "features", "pearson").head()[0]
```

```python
from pyspark.sql import functions as F

# 1. Aggregate basic stats per state
state_summary = combined_df.groupBy("zip_state").agg(
    F.count("user_id").alias("num_reviews"),
    F.countDistinct("gmap_id").alias("num_businesses"),
    F.avg("rating").alias("avg_rating")
).withColumn(
    "avg_reviews_per_store", F.col("num_reviews") / F.col("num_businesses")
)

# 2. Compute reviews per store
reviews_per_store_df = combined_df.groupBy("zip_state", "gmap_id").agg(
    F.count("user_id").alias("reviews_per_store")
)

# 3. Compute detailed quantiles per state
detailed_quantiles = [0.1, 0.25, 0.5, 0.75, 0.9]  # 10th, 25th, median, 75th, 90th percentiles

quantile_df = reviews_per_store_df.groupBy("zip_state").agg(
    F.min("reviews_per_store").alias("min_reviews_per_store"),
    F.expr(f"percentile_approx(reviews_per_store, {detailed_quantiles[0]})").alias("q10_reviews_per_store"),
    F.expr(f"percentile_approx(reviews_per_store, {detailed_quantiles[1]})").alias("q25_reviews_per_store"),
    F.expr(f"percentile_approx(reviews_per_store, {detailed_quantiles[2]})").alias("median_reviews_per_store"),
    F.expr(f"percentile_approx(reviews_per_store, {detailed_quantiles[3]})").alias("q75_reviews_per_store"),
    F.expr(f"percentile_approx(reviews_per_store, {detailed_quantiles[4]})").alias("q90_reviews_per_store"),
    F.max("reviews_per_store").alias("max_reviews_per_store")
)

# 4. Join with main summary
state_summary = state_summary.join(quantile_df, on="zip_state", how="left")

# 5. Round numeric columns to 2 decimals
cols_to_round = [
    "avg_rating", "avg_reviews_per_store",
    "min_reviews_per_store", "q10_reviews_per_store", "q25_reviews_per_store",
    "median_reviews_per_store", "q75_reviews_per_store", "q90_reviews_per_store", "max_reviews_per_store"
]

for col in cols_to_round:
    state_summary = state_summary.withColumn(col, F.round(F.col(col), 2))

# 6. Sort by average reviews per store
state_summary = state_summary.orderBy("avg_reviews_per_store", ascending=False)

state_summary.show(truncate=False)
```

#### EDA1: Business Closure Patterns

```python
# Closure rates by state
from pyspark.sql.functions import count, sum, col

closure_by_state = meta_df.groupBy("zip_state").agg(
    count("gmap_id").alias("num_businesses"),
    sum((col("state") == "Permanently closed").cast("int")).alias("num_closed")
)

closure_by_state = closure_by_state.withColumn(
    "closure_rate", col("num_closed") / col("num_businesses")
)
closure_by_state.orderBy("closure_rate", ascending=False).show()
```

```python
from pyspark.sql.functions import col, count, sum, coalesce, lit

closure_by_city = (
    meta_df
    .groupBy("zip_city")
    .agg(
        count("gmap_id").alias("num_businesses"),
        coalesce(
            sum(col("state").like("%Permanently closed%").cast("int")),
            lit(0)
        ).alias("num_closed")
    )
    .withColumn("closure_rate", col("num_closed") / col("num_businesses"))
)

closure_by_city.orderBy(col("num_closed").desc(), col("closure_rate").desc()).show(25, truncate=False)
```

```python
from pyspark.sql.functions import col, count, sum, coalesce, lit

closure_by_city = (
    meta_df
    .groupBy("category")
    .agg(
        count("gmap_id").alias("num_businesses"),
        coalesce(
            sum(col("state").like("%Permanently closed%").cast("int")),
            lit(0)
        ).alias("num_closed")
    )
    .withColumn("closure_rate", col("num_closed") / col("num_businesses"))
)

closure_by_city \
    .filter(col("num_businesses") > 50) \
    .orderBy(col("closure_rate").desc()) \
    .show(25, truncate=False)
```

#### EDA2: Neighborhood / Spatial EDA

```python
from pyspark.ml.feature import VectorAssembler

vec = VectorAssembler(
    inputCols=["latitude", "longitude"],
    outputCol="features"
)

geo_df = vec.transform(combined_df)
```

```python
from pyspark.ml.clustering import KMeans

kmeans = KMeans(k=50, seed=42)  # 50 artificial neighborhoods
model = kmeans.fit(geo_df)
clustered = model.transform(geo_df)
```

```python
meta_df = meta_df.withColumn(
    "label", F.when((F.col("avg_rating") < 2.5) | (F.col("num_of_reviews") < 5), 1).otherwise(0)
)

from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml import Pipeline

assembler = VectorAssembler(
    inputCols=["avg_rating", "num_of_reviews", "latitude", "longitude"],
    outputCol="features"
)

rf = RandomForestClassifier(labelCol="label", featuresCol="features", numTrees=50)
pipeline = Pipeline(stages=[assembler, rf])
model = pipeline.fit(meta_df)
```
